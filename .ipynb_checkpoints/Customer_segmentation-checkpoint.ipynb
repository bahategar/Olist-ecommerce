```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./result/cleaned_df.csv')
print(df.info())
df.head()
```

```python
df['date'] = pd.DatetimeIndex(df['order_date']).date 
print(df.info())
df.head()
```

# RFM Approximation

**Recency**

```python
# Recency
now = df['date'].max() + pd.Timedelta(days=1)
recency = df.groupby(['customer_unique_id'])['date'].max().reset_index()
recency['Recency'] = (now - recency['date']).dt.days
recency.drop('date', axis=1, inplace=True)
print(recency.info())
recency.head()
```

```python
sns.distplot(recency.Recency , color="dodgerblue", axlabel='Recency')
plt.show()
```

```python
# find WCSS
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++')
    kmeans.fit(recency[['Recency']])
    wcss.append(kmeans.inertia_)

# plot elbow graph
plt.plot(range(1,11),wcss,marker='o');
```

```python
kmeans = KMeans(n_clusters=3, random_state=1)
kmeans.fit(recency[['Recency']])
cluster_labels = kmeans.labels_
kmeans
```

```python
recency_segmentation = recency.copy()
recency_segmentation['recencyScore'] = kmeans.predict(recency[['Recency']])
recency_segmentation.head()
```

```python
recency_segmentation.groupby('recencyScore').describe()
```

```python
# Modify Recency Score
sorted_mean_recency = recency_segmentation\
                        .groupby('recencyScore')\
                        .agg({'Recency':'mean'})\
                        .sort_values(by='Recency', ascending=True)
sorted_mean_recency['rank'] = np.arange(len(sorted_mean_recency) - 1, -1, -1)
mapper = sorted_mean_recency[['rank']].to_dict()['rank']


recency_segmentation['recencyScore'] = recency_segmentation['recencyScore'].apply(lambda x: mapper[x])

recency_segmentation.groupby('recencyScore').describe()
```

**Frequency**

```python
# Frequency
frequency = df.drop_duplicates(subset=['order_id', 'customer_unique_id'], keep='first')\
            .groupby(['customer_unique_id'])['date']\
            .count()\
            .reset_index()
frequency = frequency.rename(columns={'date': 'Frequency'})
print(frequency.info())
frequency.head()
```

```python
sns.distplot(frequency.Frequency , color="dodgerblue", axlabel='Recency')
plt.show()
```

```python
# find WCSS
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++')
    kmeans.fit(frequency[['Frequency']])
    wcss.append(kmeans.inertia_)

# plot elbow graph
plt.plot(range(1,11),wcss,marker='o');
```

```python
kmeans = KMeans(n_clusters=3, random_state=1)
kmeans.fit(frequency[['Frequency']])
cluster_labels = kmeans.labels_
kmeans
```

```python
frequency_segmentation = frequency.copy()
frequency_segmentation['frequencyScore'] = kmeans.predict(frequency[['Frequency']])
frequency_segmentation.head()
```

```python
frequency_segmentation.groupby('frequencyScore').describe()
```

**Monetary**

```python
# Monetary
monetary = df.groupby('customer_unique_id').agg({'price': 'sum'}).reset_index()
monetary = monetary.rename(columns={'price':'Monetary'})
print(monetary.info())
monetary.head()
```

```python
sns.distplot(monetary.Monetary , color="dodgerblue", axlabel='Recency')
plt.show()
```

```python
# find WCSS
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++')
    kmeans.fit(monetary[['Monetary']])
    wcss.append(kmeans.inertia_)

# plot elbow graph
plt.plot(range(1,11),wcss,marker='o');
```

```python
kmeans = KMeans(n_clusters=3, random_state=1)
kmeans.fit(monetary[['Monetary']])
cluster_labels = kmeans.labels_
kmeans
```

```python
monetary_segmentation = monetary.copy()
monetary_segmentation['monetaryScore'] = kmeans.predict(monetary[['Monetary']])
monetary_segmentation.head()
```

```python
monetary_segmentation.groupby('monetaryScore').describe()
```

**Accumulation**

```python
rfm_table = df.groupby(['customer_unique_id'])\
                .size()\
                .reset_index()\
                .drop(0, axis=1)

# Merge the data
rfm_table = rfm_table \
    .merge(recency_segmentation, on = 'customer_unique_id') \
    .merge(frequency_segmentation, on = 'customer_unique_id') \
    .merge(monetary_segmentation, on = 'customer_unique_id')

# Set CustomerID as index
rfm_table = rfm_table.set_index('customer_unique_id')

# Calculate total score
rfm_table['score'] = rfm_table['recencyScore'] + rfm_table['frequencyScore'] + rfm_table['monetaryScore']
rfm_table = rfm_table.drop(['recencyScore', 'frequencyScore', 'monetaryScore'], axis=1)

rfm_table.head()
```

```python
rfm_table.groupby('score').mean()
```

```python
# Clustering

def clustering(x):
    if x >= 5:
        return 'high value'
    if x >= 2:
        return 'mid value'
    return 'low value'

rfm_table['Cluster'] = rfm_table['score'].apply(clustering)

rfm_table.head()
```

```python
rfm_table.groupby('Cluster')[['Recency', 'Frequency', 'Monetary']].describe()
```

```python
print(rfm_table.info())
rfm_table.reset_index()['customer_unique_id'].nunique()
```

```python
df = pd.merge(df, rfm_table.reset_index(), on='customer_unique_id')

print(df.info())
df.head()
```

```python
df.to_csv('result/clustered_df.csv', index=False)
```
